```python
from audio2numpy import open_audio
from scipy.fftpack import fft
import numpy as np
import matplotlib.pyplot as plt
# import mpld3
# mpld3.enable_notebook() #uncomment for an explorative plot
gnote, sampling_rate = open_audio("./recs/Guitar_G_note.mp3")
short_gnote, sampling_rate = open_audio("./recs/Guitar_G_note_gek√ºrzt.mp3")
g_major, sampling_rate = open_audio("./recs/Guitar_G_major.mp3")
g_major_c_major, sampling_rate = open_audio("./recs/Guitar_G_major.mp3")
hi, sampling_rate = open_audio("./recs/hi.mp3")
hi_singing, sampling_rate = open_audio("./recs/hi_singing.mp3")
print("samples in gnote: ",gnote.shape," sampling rate: ",sampling_rate)
```

```python
def showDFT(signal,filename):
  dft_signal = 2*abs(fft(signal))/len(signal)
  xAxis = np.linspace(0,sampling_rate,len(signal))
  fig, ax = plt.subplots()
  plt.plot(xAxis,dft_signal)
  #making the plot look nicer
  freq_g_tones = [49,98,196,392,783.99,1567.98] #Some frequencies of the G Tone in Herz 
  ax.set_xticks(freq_g_tones)
  plt.xlim(0,790)
  plt.ylim(0)
  plt.xlabel("Frequency in Hz")
  plt.ylabel("Fourier coefficient")
  plt.savefig(filename, dpi=600)#save figures
  plt.show()
```

```python
showDFT(gnote,"./figs/gnote.png")
```

```python
showDFT(short_gnote,"./figs/short_gnote.png")
```

```python
showDFT(g_major,"./figs/g_major.png")
```

```python
showDFT(g_major_c_major,"./figs/g_major_c_major.png")
```
